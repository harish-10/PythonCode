```python
import pandas as pd
import numpy as np
```

```python
link="https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df=pd.read_csv(link,header=None)
```

```python
headers=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns=headers
```

```python
df.head()
```

```python
df.replace('?',np.NaN,inplace=True)
```

```python
df.dropna(subset=["price"], axis=0,inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df['normalized-losses']=df['normalized-losses'].astype(float)
```

```python
df
```

```python
df['normalized-losses']=df['normalized-losses'].replace(np.NaN,df['normalized-losses'].mean())
```

```python
df.isnull().any()
```

```python
df['num-of-doors']=df['num-of-doors'].replace(np.NaN,df['num-of-doors'].value_counts().idxmax())
```

```python
df['bore']=df['bore'].astype(float)
df['stroke']=df['stroke'].astype(float)
df['peak-rpm']=df['peak-rpm'].astype(float)
df['horsepower']=df['horsepower'].astype(float)
df['price']=df['price'].astype(int)
```

```python
df['bore']=df['bore'].replace(np.NaN,df['bore'].mean())
df['stroke']=df['stroke'].replace(np.NaN,df['stroke'].mean())
df['horsepower']=df['horsepower'].replace(np.NaN,df['horsepower'].mean())
df['peak-rpm']=df['peak-rpm'].replace(np.NaN,df['peak-rpm'].mean())
```

```python
df.head()
```

```python
df['city-L/100km'] = 235/df["city-mpg"]
df['highway-L/100km'] = 235/df["highway-mpg"]
df.drop(['city-mpg','highway-mpg'],axis=1)
```

```python
df['normalized-losses']=(df['normalized-losses']-df['normalized-losses'].min())/(df['normalized-losses'].max()-df['normalized-losses'].min())
df['wheel-base']=(df['wheel-base']-df['wheel-base'].min())/(df['wheel-base'].max()-df['wheel-base'].min())
df['engine-size']=(df['engine-size']-df['engine-size'].min())/(df['engine-size'].max()-df['engine-size'].min())
df['bore']=(df['bore']-df['bore'].min())/(df['bore'].max()-df['bore'].min())
df['stroke']=(df['stroke']-df['stroke'].min())/(df['stroke'].max()-df['stroke'].min())
df['compression-ratio']=(df['compression-ratio']-df['compression-ratio'].min())/(df['compression-ratio'].max()-df['compression-ratio'].min())
df['horsepower']=(df['horsepower']-df['horsepower'].min())/(df['horsepower'].max()-df['horsepower'].min())
df['peak-rpm']=(df['peak-rpm']-df['peak-rpm'].min())/(df['peak-rpm'].max()-df['peak-rpm'].min())
df['city-L/100km']=(df['city-L/100km']-df['city-L/100km'].min())/(df['city-L/100km'].max()-df['city-L/100km'].min())
df['highway-L/100km']=(df['highway-L/100km']-df['highway-L/100km'].min())/(df['highway-L/100km'].max()-df['highway-L/100km'].min())
df['length']=(df['length']-df['length'].min())/(df['length'].max()-df['length'].min())
df['width']=(df['width']-df['width'].min())/(df['width'].max()-df['width'].min())
df['height']=(df['height']-df['height'].min())/(df['height'].max()-df['height'].min())
df['curb-weight']=(df['curb-weight']-df['curb-weight'].min())/(df['curb-weight'].max()-df['curb-weight'].min())



```

```python
df.describe(include="all")
```

```python
binwidth=int((max(df["price"])-min(df["price"]))/4)
```

```python
bins=np.linspace(min(df["price"]),max(df["price"]),4)
```

```python
group_names=['Low','Medium','High']
df['price_binned']=pd.cut(df['price'],bins,labels=group_names)
```

```python
df.head()
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
# Write your code below and press Shift+Enter to execute 
dummy_variable_2 = pd.get_dummies(df["aspiration"])
dummy_variable_2.rename(columns={'aspiration-type-std':'std', 'aspiration-type-turbo':'turbo'}, inplace=True)
# Write your code below and press Shift+Enter to execute 
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_2], axis=1)

# drop original column "fuel-type" from "df"
df.drop("aspiration", axis = 1, inplace=True)
```
